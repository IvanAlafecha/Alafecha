```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder

# Cargar el dataframe desde el archivo parquet
df = pd.read_parquet('train202212_202307.parquet')

# Selección de las variables requeridas para entrenar el modelo
X = df[['DiasAtraso', 'TipoPrestamo', 'Relacion_coutasPagasPlazo', 'Relacion_pagoTotal', 'Genero', 
        'TipoVivienda', 'LocalizacionNegocio', 'Edad', 'EstadoCivil', 'Hijos', 'PersonasCargo', 
        'Calificacion_ult_gestion', 'Calificacion_ult_gestion_3_meses', 'Calificacion_ult_gestion_6_meses', 
        'Recuento Incumplido 3 meses', 'Recuento Incumplido 6 meses', 'Recuento con contacto ultimos 3 meses', 
        'Recuento con contacto ultimos 6 meses', 'Recuento_sin_contacto_ultimos_3_meses', 
        'Recuento_sin_contacto_ultimos_6_meses']]

# Selección de la variable objetivo
y = df["P_recuperacionF"]

# Lista de las columnas categóricas para las que se aplicará Label Encoding
categorical_columns = ['TipoPrestamo', 'Genero', 'TipoVivienda', 'LocalizacionNegocio', 'EstadoCivil']

# Inicializar el LabelEncoder
label_encoder = LabelEncoder()

# Aplicar Label Encoding a las columnas categóricas
for column in categorical_columns:
    X[column] = label_encoder.fit_transform(X[column])

# Verificar cómo queda el dataframe después de Label Encoding
print(X.head())

# Si necesitas realizar el One-Hot Encoding en el futuro, puedes desactivar esta parte y volver a aplicar pd.get_dummies()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Calculamos la matriz de correlación
correlation_matrix = X_encoded.corr()

# Aumentar el tamaño de la figura y ajustar los tamaños de las celdas
plt.figure(figsize=(18, 14))  # Tamaño de la figura (ancho, alto)
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5, 
            annot_kws={"size": 10},  # Tamaño de las anotaciones dentro de las celdas
            cbar_kws={"shrink": 0.8},  # Reduce el tamaño de la barra de colores
            square=True,  # Mantener la matriz cuadrada (si tienes la misma cantidad de filas y columnas)
            xticklabels=correlation_matrix.columns,  # Etiquetas en el eje X
            yticklabels=correlation_matrix.columns)  # Etiquetas en el eje Y

# Rotar las etiquetas de los ejes X y Y para que se vean correctamente
plt.xticks(rotation=90, ha='right', fontsize=10)  # Rotar las etiquetas X
plt.yticks(rotation=0, va='center', fontsize=10)  # Asegurarnos que las etiquetas Y estén centradas

# Título de la gráfica
plt.title('Matriz de Correlación', fontsize=16)

# Mostrar la gráfica
plt.tight_layout()  # Ajusta todo para que no se sobrepongan
plt.show()
```

```python
# Importación de librerías
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from sklearn.preprocessing import StandardScaler

# Escalar las características numéricas
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_encoded)

# Dividir el conjunto de datos en entrenamiento y prueba
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

# Crear y entrenar el modelo de Random Forest
rf_model = RandomForestRegressor(n_estimators=100, random_state=42)
rf_model.fit(X_train, y_train)

# Realizar predicciones
y_pred_rf = rf_model.predict(X_test)

# Evaluar el modelo
rmse_rf = np.sqrt(mean_squared_error(y_test, y_pred_rf))
mae_rf = mean_absolute_error(y_test, y_pred_rf)
r2_rf = r2_score(y_test, y_pred_rf)

# Calcular el MAPE con epsilon para evitar problemas con valores cercanos a cero
epsilon = 1e-8  # Valor pequeño para evitar división por cero
mape_rf = np.mean(np.abs((y_test - y_pred_rf) / np.maximum(np.abs(y_test), epsilon))) * 100

# Imprimir los resultados de las métricas
print(f"RMSE del modelo Random Forest: {rmse_rf:.4f}")
print(f"MAE del modelo Random Forest: {mae_rf:.4f}")
print(f"R2 del modelo Random Forest: {r2_rf:.4f}")
print(f"MAPE del modelo Random Forest: {mape_rf:.2f}%")
```

```python
# Importación de librerías
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split

# Escalar las características numéricas
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_encoded)

# Dividir el conjunto de datos en entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

# Crear y entrenar el modelo de Random Forest
rf_model = RandomForestRegressor(n_estimators=100, random_state=42)
rf_model.fit(X_train, y_train)

# Realizar predicciones para entrenamiento y prueba
y_pred_train = rf_model.predict(X_train)
y_pred_test = rf_model.predict(X_test)

# Evaluar el modelo en el conjunto de entrenamiento
rmse_train = np.sqrt(mean_squared_error(y_train, y_pred_train))
mae_train = mean_absolute_error(y_train, y_pred_train)
r2_train = r2_score(y_train, y_pred_train)

# Evaluar el modelo en el conjunto de prueba
rmse_test = np.sqrt(mean_squared_error(y_test, y_pred_test))
mae_test = mean_absolute_error(y_test, y_pred_test)
r2_test = r2_score(y_test, y_pred_test)

# Calcular el MAPE con epsilon para evitar problemas con valores cercanos a cero
epsilon = 1e-8  # Valor pequeño para evitar división por cero
mape_train = np.mean(np.abs((y_train - y_pred_train) / np.maximum(np.abs(y_train), epsilon))) * 100
mape_test = np.mean(np.abs((y_test - y_pred_test) / np.maximum(np.abs(y_test), epsilon))) * 100

# Imprimir los resultados de las métricas para entrenamiento y prueba
print(f"Evaluación en entrenamiento:")
print(f"RMSE: {rmse_train:.4f}")
print(f"MAE: {mae_train:.4f}")
print(f"R2: {r2_train:.4f}")
print(f"MAPE: {mape_train:.2f}%\n")

print(f"Evaluación en prueba:")
print(f"RMSE: {rmse_test:.4f}")
print(f"MAE: {mae_test:.4f}")
print(f"R2: {r2_test:.4f}")
print(f"MAPE: {mape_test:.2f}%")
```
