```python
import pandas as pd

# Suponiendo que ya tienes un DataFrame llamado 'Pagos_marcados'

# Asegurarse de que la columna 'Fecha Pago' esté en formato datetime
Pagos_marcados['Fecha Pago'] = pd.to_datetime(Pagos_marcados['Fecha Pago'], errors='coerce')

# Eliminar filas con Fecha Pagos inválidas (NaT)
Pagos_marcados = Pagos_marcados.dropna(subset=['Fecha Pago'])

# Extraer el mes y el año de la Fecha Pago para facilitar la agrupación
Pagos_marcados['Mes_Año'] = Pagos_marcados['Fecha Pago'].dt.to_period('M')

# Agrupar por 'CREDITO', 'DOCUMENTO' y 'Mes_Año' y sumar los pagos de 'TOTAL CANCELADO'
agrupado = Pagos_marcados.groupby(['CREDITO', 'DOCUMENTO', 'Mes_Año'])['TOTAL CANCELADO'].sum().reset_index()

# Crear una lista de todos los meses desde Enero 2021 hasta Diciembre 2024
meses = pd.date_range(start='2021-01-01', end='2024-12-01', freq='MS').strftime('%b-%Y')

# Crear un DataFrame vacío para el resultado final
# Crear las combinaciones de 'CREDITO' y 'DOCUMENTO'
combinaciones = agrupado[['CREDITO', 'DOCUMENTO']].drop_duplicates()

# Crear las columnas de 'CREDITO', 'DOCUMENTO' y los meses
new_df = pd.DataFrame(combinaciones)

# Inicializamos las columnas de los meses con 0
for mes in meses:
    new_df[mes] = 0

# Rellenar las columnas de meses con los pagos correspondientes
for index, row in agrupado.iterrows():
    documento = row['DOCUMENTO']
    mes_año = row['Mes_Año']
    total_cancelado = row['TOTAL CANCELADO']
    credito = row['CREDITO']

    # Convertir el mes_año a formato 'Mes-Año'
    nombre_mes = mes_año.strftime('%b-%Y')

    # Encontrar el índice del documento y crédito en el DataFrame final
    idx = new_df[(new_df['CREDITO'] == credito) & (new_df['DOCUMENTO'] == documento)].index[0]
    
    # Asignar el total_cancelado al mes correspondiente
    new_df.at[idx, nombre_mes] += total_cancelado

# Ver el nuevo DataFrame
print(new_df.head())

```

```python
import pandas as pd

# Asegurarse de que la columna 'Fecha Marcado' esté en formato datetime
D_marcados_nuevo['Fecha Marcado'] = pd.to_datetime(D_marcados_nuevo['Fecha Marcado'], errors='coerce')

# Crear una lista de los meses desde Ene-2021 hasta Dic-2024
meses = pd.date_range(start='2021-01-01', end='2024-12-01', freq='MS').strftime('%b-%Y')

# Agregar una columna 'Mes_Año' al DataFrame con el formato de mes-año
D_marcados_nuevo['Mes_Año'] = D_marcados_nuevo['Fecha Marcado'].dt.to_period('M')

# Agrupar por 'Mes_Año' y calcular la suma de los pagos y saldo de capital
agrupado = D_marcados_nuevo.groupby('Mes_Año')[meses].sum()  # Sumar los pagos por mes
saldo_capital_agrupado = D_marcados_nuevo.groupby('Mes_Año')['Saldo Capital'].sum()  # Sumar el saldo capital por mes

# Crear un DataFrame vacío con las filas y columnas correspondientes a los meses
new_df = pd.DataFrame(index=meses, columns=meses)

# Iterar sobre cada mes para llenar el DataFrame 'new_df'
for i, mes in enumerate(meses):
    # Para cada fila (mes), calculamos los valores acumulados hasta ese mes
    pagos_acumulados = 0
    saldo_capital_acumulado = 0
    
    for j, mes2 in enumerate(meses):
        # Si el mes 'mes' es anterior o igual a 'mes2', calculamos el porcentaje acumulado
        if pd.to_datetime(mes) <= pd.to_datetime(mes2):
            # Acumulamos los pagos hasta el mes 'mes2'
            pagos_acumulados += agrupado.loc[:pd.to_datetime(mes2).to_period('M')].iloc[:, i].sum()
            saldo_capital_acumulado = saldo_capital_agrupado.loc[:pd.to_datetime(mes2).to_period('M')].sum()

            # Calculamos el porcentaje acumulado
            if saldo_capital_acumulado > 0:
                new_df.loc[mes, mes2] = pagos_acumulados / saldo_capital_acumulado
            else:
                new_df.loc[mes, mes2] = 0  # Si el saldo capital es 0, asignar 0 a la celda

# Ver el DataFrame final
print(new_df)
```
